## Constant Time: $\mathcal{O}(1)$
- Generally constant time or $\mathcal{O}(1)$ means that it takes a ***fixed*** number of instructions to execute a piece of code. And each time you run the same piece of code, it will always run for a constant amount of time.
- However, it is often confused for running something quickly, which is not always true. $\mathcal{O}(1)$ does not necessarily mean "quickly". It means that the time it takes is constant, and not based on the size of the input to the function. Constant could be fast or slow. 

```python
def constant_algo(items):
    loops = 0
    result = items[0] * items[0]
    loops+=1
    return loops

constant_algo([4, 5, 6, 8])
```

## Linear Time: $\mathcal{O}(N)$
- $\mathcal{O}(N)$ means that the time taken will change linearly with respect to the size of the input to the function, denoted by $N$. Again, it could be fast or slow, but it will get slower as the size of $N$ increases.
- For example, if running a linear algorithm on input of size N=100 takes, say, 1 minute, then running the same algorithm on input of size N=200 would take about 2 minutes. 

```python
def linear_algo(items):
    loops=0
    for item in items:
        a = item + 1
        loops+=1
    return loops

linear_algo([4, 5, 6, 8])
```

## Quadratic Time $\mathcal{O}(N^2)$
- Similar to linear time, an algorithm taking quadratic time means that it is dependent on the input size $N$, but unlike linear time, this will increase more rapidly as $N$ increases.
- For example, if running a quadratic algorithm on input of size $N = 100$ takes, say, 1 minute, then running the same algorithm on input of size $N = 200$ would take about 4 minutes.

```python
def quadratic_algo(items):
    loops=0
    for item in items:
        for item2 in items:
            a = item + item2
            loops+=1
    return loops

quadratic_algo([4, 5, 6, 8])
```

## Logarithmic Time $\mathcal{O}(logN)$
- One way to describe logarithmic time $\mathcal{O}(log N)$ is when time goes up linearly when $N$ goes up exponentially.
- For instance, if it takes 1 second to compute an array of size $N=10$, it will take 2 seconds to compute $N=100$, 3 seconds to compute $N=1000$ elements, and so on.
- In this example, it is $\mathcal{O}(log_{10}(N))$, but this way of describing time complexity works for any $log_x$ with base $x$.

```python
def log_algo(items):
    loops=0
    i = 1
    while i < len(items):
        a = items[i] + 1
        i = i * 2
        loops+=1
    return loops

log_algo([4, 5, 6, 8])
```

## Empirically Evaluate Time Complexity
- Below is a short code snippet to evaluate the above example algorithms. 
- Plotting the time complexity will give you an idea of how long each of them will take to run as you increase the size of the array.

```python
import random
import timeit
from functools import partial
from matplotlib import pyplot
def plot_time_complexity(fn, nMin, nMax, nInc, nTests):
    # Run timer and plot time complexity
    x = []
    y = []
    for i in range(nMin, nMax, nInc):
        N = i
        testNTimer = timeit.Timer(partial(fn, random.sample(range(0, nMax), N-1)))
        t = testNTimer.timeit(number=nTests)
        x.append(i)
        y.append(t)
    p1 = pyplot.plot(x, y, 'o')
```

```python
plot_time_complexity(linear_algo, 100, 5000, 100, 10)
```

```python
plot_time_complexity(quadratic_algo, 10, 500, 10, 10)
```

```python
plot_time_complexity(log_algo, 10, 5000, 10, 10)
```

## Problem 3.1:
- The code below is using linear search to find an element `x` in the list `arr`.
- It will sequentially (one by one, starting with the first element in `arr`) look for each element in the list `arr` until a match is found or the whole list has been searched.
- Question: What is its worst-case time complexity?
***

```python
def linear_search(arr, x): 
    for i in range(len(arr)): 
  
        if arr[i] == x: 
            return i 
  
    return -1
```

## Problem 3.2:
- The code below is another way to search for an element `x` in the list `arr`, called *Binary Search*. *Binary Search* takes in a **sorted** array and repeatedly sub-divides the search interval in half. 
- Since the array comes in sorted (lets say ascending order), you can eliminate half the numbers each time if you just compare `x` with the middle element. If the middle element happens to be smaller than `x` then every element to left will also be small (remember `arr` is sorted).
  - So we don't have to check with half the elements in the list and only look at the other half.
- This algorithm repeatedly makes these decisions until you either find the element you are looking for or if can no longer divide the list, in which case the element does not exist.
- Question: What is its worse-case time complexity?
  - (HINT) Think about how half the elements are eliminated each time. Where have we seen that before?

```python
def binary_search(arr, x): 
    low = 0
    high = len(arr) - 1
    mid = 0
  
    while low <= high: 
  
        mid = (high + low) // 2
  
        # Check if x is present at mid 
        if arr[mid] < x: 
            low = mid + 1
  
        # If x is greater, ignore left half 
        elif arr[mid] > x: 
            high = mid - 1
  
        # If x is smaller, ignore right half 
        else: 
            return mid 
  
    # If we reach here, then the element was not present 
    return -1
```

## Timing Problem 3.1
- You can look at the actual timing of the above problem to see how it grows in terms of time complexity as we increase the array size.
***

```python
import random
import timeit
from functools import partial
from matplotlib import pyplot

def plot_time_complexity_worst_case(func, nMin, nMax, nInc, nTests):
    x = []
    y = []
    for N in range(nMin, nMax, nInc):
        # Make an array of values, and search for a value not in that array.
        arr = random.sample(range(1, 10000), N) # This is the array that you will be searching in.
        val = 4999.4 # This is a value you will be searching for. It will never exist in the array.

        testNTimer = timeit.Timer(partial(func, arr, val)) # Call the function you passed in, with the array and value to find.
        t = testNTimer.timeit(number=nTests)
        x.append(N)
        y.append(t)
 
    p1 = pyplot.plot(x, y, 'o')
```

```python
# Helper function - get a sample of the rows and a set of attributes and see how long it takes to plot.
def plot_some_records(df, num_rows_to_plot, list_of_attribs_to_plot):
    sample_df = df.sample(n=num_rows_to_plot, replace=False, random_state=1)
    
    start_time = time.perf_counter()
    sns.pairplot(data = sample_df[list_of_attribs_to_plot], hue='species', height=3)
    plt.show()
    elapsed_time = time.perf_counter() - start_time

    num_attribs_to_plot = len(list_of_attribs_to_plot)-1  #ignore the species since it is the hue
    print('num_rows:', num_rows_to_plot, '   num_attribs_to_plot:', num_attribs_to_plot, '   elapsed_time:', elapsed_time)
    return num_rows_to_plot, num_attribs_to_plot, elapsed_time
```
