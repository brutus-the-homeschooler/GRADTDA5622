```python
#Create a box plot of new deaths per million of 5 most populated countries.
plt.figure(figsize=(20,10))
iso_codes_5pop = ['CHN', 'IND', 'USA', 'IDN', 'PAK']
top_5 = covid_data_df.loc[covid_data_df['iso_code'].isin(iso_codes_5pop)]
top_5 = top_5[top_5['iso_code'].notna()] #trim nan from iso_code
sb.boxplot(y=covid_data_df['new_deaths_smoothed_per_million'], x=top_5['iso_code'], hue=top_5['iso_code'])
plt.title("Box Plot of new deaths per million for the 5 most populated countries")
```

**Discussion:**
**Why the visualization is interesting:** I chose to create a boxplot of the new deaths per million based on the 5 most populated countries. I thought it would be interesting to compare the distrubtion of new deaths per million between them, and see if there was any large differences. From the graph above it is apparent that the USA, which is the 3rd most populated country in the world, is experiencing more new deaths per million than the other most populated countries. This is evident as the USA has the highest maximum, median, and a larger interquartile range. Where as the other 4 most populated countries, China, India, Indonesia, and Pakistan all have relatively similar boxplots.

```python
#Create a box plot of new tests per thousand of 5 most populated countries.
plt.figure(figsize=(20,10))
sb.boxplot(y=covid_data_df['new_tests_smoothed_per_thousand'], x=top_5['iso_code'], hue=top_5['iso_code'])
plt.title("Box Plot of new tests per thousand for the 5 most populated countries")
```

**Discussion:**
**Why this visualization is interesting:** I thought after visualizng new deaths per million, it would be interesting to see how the testing per thousand looked for the 5 most populated countries.
